```python
import scipy.stats as stats

# Sample data information
mean = 50
std_dev = 5
sample_size = 15  # Please provide the sample size

# Calculate the standard error
std_error = std_dev / (sample_size ** 0.5)

# Calculate the margin of error
margin_of_error = stats.t.ppf(0.975, df=sample_size - 1) * std_error

# Calculate the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Print the confidence interval
print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
```

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
observed_frequencies = np.array([35, 30, 15, 10, 20])  # Replace with your observed frequencies
expected_frequencies = np.array([0.2, 0.2, 0.1, 0.1, 0.2]) * sum(observed_frequencies)
```

```python
chi2, p_value = chi2_contingency(observed_frequencies, f_exp=expected_frequencies)
```

```python
df = len(observed_frequencies) - 1
alpha = 0.05
critical_value = chi2_contingency.ppf(1 - alpha, df)
```

```python
if chi2 > critical_value or p_value < alpha:
    print("Reject the null hypothesis. The distribution of colors in the M&M bag is not as expected.")
else:
    print("Fail to reject the null hypothesis. The distribution of colors in the M&M bag matches the expected distribution.")
```

```python
import math
from scipy.stats import norm

sample_size = 500
sample_proportion = 60 / sample_size
confidence_level = 0.95

z_score = norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error

confidence_interval = (lower_bound, upper_bound)

print("Confidence Interval:", confidence_interval)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the degrees of freedom
df = 10

# Generate x values for the chi-square distribution
x = np.linspace(0, 30, 500)

# Calculate the corresponding y values for the chi-square distribution
y = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, label='Chi-square (10 degrees of freedom)')

# Shade the area corresponding to chi-square statistic of 15
x_shade = np.linspace(15, 30, 500)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='grey', alpha=0.5)

# Label the axes
plt.xlabel('X')
plt.ylabel('Probability Density')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
To conduct a chi-square goodness of fit test, we need to follow these steps:

Step 1: State the hypotheses:

Null hypothesis (H₀): The coin is fair (not biased towards tails).
Alternative hypothesis (H₁): The coin is biased towards tails.
Step 2: Set the significance level:
The significance level is given as 0.05, which means we have a 5% chance of rejecting the null hypothesis when
it is actually true.

Step 3: Set up the expected frequencies:
For a fair coin, we expect an equal probability of getting heads or tails. So, for 100 flips, we would expect
50 tails and 50 heads.

Step 4: Calculate the test statistic:
The test statistic for a chi-square goodness of fit test is calculated using the formula:
χ² = ∑((Oᵢ - Eᵢ)² / Eᵢ)
where Oᵢ is the observed frequency and Eᵢ is the expected frequency.

In this case, we have:
Oᵢ = 45 tails
Eᵢ = 50 tails

χ² = ((45 - 50)² / 50) = 0.5

Step 5: Determine the critical value:
The critical value for a chi-square test with one degree of freedom (since we are comparing observed frequencies
to expected frequencies) at a significance level of 0.05 is approximately 3.841.

Step 6: Compare the test statistic with the critical value:
Since our test statistic (0.5) is less than the critical value (3.841), we fail to reject the null hypothesis.

Step 7: Interpret the result:
Based on the chi-square goodness of fit test, there is not enough evidence to conclude that the observed
frequencies significantly differ from the expected frequencies of a fair coin. Therefore, we do not have
enough evidence to support the claim that the coin is biased towards tails.

```

```python
To conduct a hypothesis test to determine if the population mean is significantly different from 70, 
we can use a one-sample t-test. Here are the steps to perform the hypothesis test:

Step 1: State the null hypothesis (H0) and the alternative hypothesis (H1):

Null hypothesis (H0): The population mean is equal to 70 (µ = 70).
Alternative hypothesis (H1): The population mean is significantly different from 70 (µ ≠ 70).
Step 2: Determine the significance level (α):

The given significance level is 0.05, which corresponds to a 5% level of significance.
Step 3: Calculate the test statistic:

The test statistic for a one-sample t-test is given by: t = (sample mean - hypothesized mean) / 
(sample standard deviation / √sample size)
Given data:
Sample mean (x̄) = 72
Hypothesized mean (µ) = 70
Sample standard deviation (s) = 10
Sample size (n) = 30
Calculate the test statistic:
t = (72 - 70) / (10 / √30) ≈ 2.108
Step 4: Determine the critical value(s):

Since we have a two-tailed test (µ ≠ 70), we need to consider both tails of the t-distribution.
With a significance level of 0.05 and 30 degrees of freedom (n - 1 = 30 - 1 = 29), we can find the
critical value(s) using a t-table or a statistical software. For a two-tailed test, each tail will have an area of 0.025.
Using a t-table, the critical value for α/2 = 0.025 and 29 degrees of freedom is approximately 2.045.
Step 5: Compare the test statistic with the critical value(s):

Since the absolute value of the test statistic (|2.108| = 2.108) is greater than the critical value (2.045),
we reject the null hypothesis.
Step 6: Make a conclusion:

Based on the results of the hypothesis test, there is sufficient evidence to conclude that the population mean is
significantly different from 70 at a significance level of 0.05.
Therefore, we reject the null hypothesis and conclude that the population mean is significantly different from 70.

```
